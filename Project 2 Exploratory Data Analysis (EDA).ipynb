### [C4] For each customer, what proportion of their products ordered are reorders?

```python
# Merge the data on order_id
merged_data = pd.merge(order_products_df, orders_df, on='order_id')

# Group by user_id and calculate the total number of products ordered and reorders
user_data = merged_data.groupby('user_id').agg({
    'product_id': 'count',
    'reordered': 'sum'
}).reset_index()

# Rename the columns
user_data.columns = ['user_id', 'total_products', 'reorders']

# Calculate the proportion of reorders
user_data['proportion_reorders'] = user_data['reorders'] / user_data['total_products']

# Print the result
print(user_data)
```

```python
mean_proportion_reorders = user_data['proportion_reorders'].mean()
print(mean_proportion_reorders)
```

```python
plt.hist(user_data['proportion_reorders'], bins=20, alpha=0.5, label='Proportion of Reorders')

# Add a title and labels
plt.title('Distribution of Proportion of Reorders')
plt.xlabel('Proportion of Reorders')
plt.ylabel('Frequency')

# Show the mean proportion of reorders as a vertical line
plt.axvline(mean_proportion_reorders, color='r', linestyle='dashed', label='Mean')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
